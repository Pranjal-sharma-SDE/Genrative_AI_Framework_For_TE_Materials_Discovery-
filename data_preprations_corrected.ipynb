```python
# ============================================================
# FULL 118 ELEMENT LIST
# ============================================================

ELEMENTS = [
    "H", "He",
    "Li", "Be", "B", "C", "N", "O", "F", "Ne",
    "Na", "Mg", "Al", "Si", "P", "S", "Cl", "Ar",
    "K", "Ca", "Sc", "Ti", "V", "Cr", "Mn", "Fe", "Co", "Ni",
    "Cu", "Zn", "Ga", "Ge", "As", "Se", "Br", "Kr",
    "Rb", "Sr", "Y", "Zr", "Nb", "Mo", "Tc", "Ru", "Rh",
    "Pd", "Ag", "Cd", "In", "Sn", "Sb", "Te", "I", "Xe",
    "Cs", "Ba",
    "La", "Ce", "Pr", "Nd", "Pm", "Sm", "Eu", "Gd",
    "Tb", "Dy", "Ho", "Er", "Tm", "Yb", "Lu",
    "Hf", "Ta", "W", "Re", "Os", "Ir", "Pt", "Au", "Hg",
    "Tl", "Pb", "Bi", "Po", "At", "Rn",
    "Fr", "Ra",
    "Ac", "Th", "Pa", "U", "Np", "Pu", "Am", "Cm",
    "Bk", "Cf", "Es", "Fm", "Md", "No", "Lr",
    "Rf", "Db", "Sg", "Bh", "Hs", "Mt", "Ds", "Rg",
    "Cn", "Nh", "Fl", "Mc", "Lv", "Ts", "Og"
]
```

```python


# ============================================================
# VALID ELEMENTS
# ============================================================

VALID_ELEMENT_SET = set(
    el.symbol for el in Element
)

# ============================================================
# TOKEN PATTERN
# ============================================================

TOKEN_PATTERN = re.compile(
    r'([A-Z][a-z]?)([0-9]*\.?[0-9]*)'
)

# ============================================================
# REMOVE TAGS
# ============================================================

REMOVE_TAGS = [
    "HP",
    "HPCS",
    "SPS",
    "SPS1100",
    "annealed",
    "double filled",
    "single filled",
    "melt spun",
    "hot pressed",
    "hot-pressed",
    "arc melted",
    "quenched",
    "aged",
    "cast",
    "polycrystalline",
    "nanostructured",
    "bulk",
    "film",
    "pellet",
    "sample",
    "engineering ZT",
    "p-type",
    "n-type",
]

# ============================================================
# STRICT FORMULA CHECK
# ============================================================

def strict_formula_check(formula):

    """
    Ensures:
    1. Entire string is valid
    2. No fake elements
    3. No leftover characters
    """

    formula = str(formula).strip()

    if formula == "":
        return False

    matches = list(TOKEN_PATTERN.finditer(formula))

    if not matches:
        return False

    reconstructed = ""

    for m in matches:

        elem = m.group(1)
        amt = m.group(2)

        # ----------------------------------------
        # INVALID ELEMENT
        # ----------------------------------------

        if elem not in VALID_ELEMENT_SET:
            return False

        reconstructed += elem + amt

    # --------------------------------------------
    # ENTIRE STRING MUST MATCH
    # --------------------------------------------

    if reconstructed != formula:
        return False

    # --------------------------------------------
    # FINAL PYMATGEN VALIDATION
    # --------------------------------------------

    try:
        Composition(formula)
        return True
    except:
        return False

# ============================================================
# EXTRACT VALID FORMULA
# ============================================================

def extract_main_formula(text):

    """
    Extracts the FIRST valid chemical formula
    from noisy text.
    """

    text = str(text)

    # Examples handled:
    # Cu2ZnSnS4(CZTS)
    # Mg2Sn0.93Pb0.05Sb0.02 (engineering ZT)
    # Bi2Te3 - annealed

    candidates = re.findall(
        r'[A-Z][A-Za-z0-9\.\(\)]*',
        text
    )

    candidates = sorted(
        candidates,
        key=len,
        reverse=True
    )

    for cand in candidates:

        cand = cand.strip()

        # ----------------------------------------
        # REMOVE PARENTHESES CONTENT
        # ----------------------------------------

        cand = re.sub(r'\(.*?\)', '', cand)

        # ----------------------------------------
        # REMOVE SPACES
        # ----------------------------------------

        cand = re.sub(r'\s+', '', cand)

        # ----------------------------------------
        # STRICT VALIDATION
        # ----------------------------------------

        if strict_formula_check(cand):
            return cand

    return None

# ============================================================
# CLEAN FORMULA
# ============================================================

def clean_formula(text):

    text = str(text).strip()

    # ========================================================
    # REMOVE GREEK PREFIXES
    # ========================================================

    text = re.sub(
        r'^[αβγδλωκ]\s*[-–]\s*',
        '',
        text
    )

    # ========================================================
    # REMOVE p-/n-
    # ========================================================

    text = re.sub(
        r'^[pn]\s*[-–]\s*',
        '',
        text,
        flags=re.IGNORECASE
    )

    # ========================================================
    # HANDLE x,y,z VARIABLES
    # ========================================================

    assignments = dict(
        re.findall(
            r'([xyz])\s*=\s*([0-9]*\.?[0-9]+)',
            text,
            flags=re.IGNORECASE
        )
    )

    for var, val in assignments.items():

        text = re.sub(
            rf'\b{var}\b',
            val,
            text,
            flags=re.IGNORECASE
        )

    # Replace remaining x/y/z

    text = re.sub(
        r'\bx\b',
        '0.05',
        text,
        flags=re.IGNORECASE
    )

    text = re.sub(
        r'\by\b',
        '0.05',
        text,
        flags=re.IGNORECASE
    )

    text = re.sub(
        r'\bz\b',
        '0.05',
        text,
        flags=re.IGNORECASE
    )

    # ========================================================
    # REMOVE TAGS
    # ========================================================

    for tag in REMOVE_TAGS:

        text = re.sub(
            re.escape(tag),
            '',
            text,
            flags=re.IGNORECASE
        )

    # ========================================================
    # REMOVE ALL PARENTHESES CONTENT
    # ========================================================

    text = re.sub(
        r'\(.*?\)',
        '',
        text
    )

    # ========================================================
    # REMOVE ADDITIVES
    # ========================================================

    text = re.split(r'[@%]', text)[0]

    # ========================================================
    # REMOVE EXTRA SYMBOLS
    # ========================================================

    text = re.sub(
        r'[^A-Za-z0-9\.]',
        '',
        text
    )

    # ========================================================
    # EXTRACT VALID FORMULA
    # ========================================================

    formula = extract_main_formula(text)

    return formula

# ============================================================
# NORMALIZE FORMULA
# ============================================================

def normalize_formula(formula):

    comp = Composition(formula)

    normalized = (
        comp.fractional_composition.formula
    )

    normalized = normalized.replace(
        " ",
        ""
    )

    return normalized

# ============================================================
# CLEAN DATASET
# ============================================================

def clean_dataset(df, formula_column="material"):

    cleaned_formulas = []
    normalized_formulas = []
    valid_indices = []

    invalid_entries = []

    for idx, item in enumerate(df[formula_column].astype(str)):

        try:

            # =================================================
            # STEP 1: CLEAN
            # =================================================

            cleaned = clean_formula(item)

            # =================================================
            # STEP 2: VALIDATE
            # =================================================

            if cleaned is None:
                invalid_entries.append(item)
                continue

            if not strict_formula_check(cleaned):
                invalid_entries.append(item)
                continue

            # =================================================
            # STEP 3: NORMALIZE
            # =================================================

            normalized = normalize_formula(cleaned)

            cleaned_formulas.append(cleaned)
            normalized_formulas.append(normalized)

            valid_indices.append(idx)

        except:
            invalid_entries.append(item)
            continue

    # ========================================================
    # BUILD CLEAN DF
    # ========================================================

    cleaned_df = df.iloc[valid_indices].copy()

    cleaned_df["Cleaned_Formula"] = cleaned_formulas

    cleaned_df["Normalized_Formula"] = normalized_formulas

    # ========================================================
    # REORDER COLUMNS
    # ========================================================

    cols = list(cleaned_df.columns)

    cols.remove("Cleaned_Formula")
    cols.remove("Normalized_Formula")

    insert_position = cols.index(formula_column) + 1

    cols.insert(insert_position, "Cleaned_Formula")
    cols.insert(insert_position + 1, "Normalized_Formula")

    cleaned_df = cleaned_df[cols]

    # ========================================================
    # PRINT SUMMARY
    # ========================================================

    print("=" * 60)
    print("DATA CLEANING SUMMARY")
    print("=" * 60)

    print(f"\nOriginal Rows: {len(df)}")
    print(f"Valid Rows:    {len(cleaned_df)}")
    print(f"Removed Rows:  {len(df) - len(cleaned_df)}")

    print("\nExamples of Removed Entries:")

    for x in invalid_entries[:20]:
        print(" -", x)

    return cleaned_df

# ============================================================
# RUN
# ============================================================



final_dataset = clean_dataset(
    zt,
    formula_column="material"
)

# ============================================================
# SAVE
# ============================================================

final_dataset.to_csv(
    "cleaned_materials_dataset.csv",
    index=False
)

# ============================================================
# RESULTS
# ============================================================

print("\n")
print("=" * 60)
print("CLEANED DATASET SAMPLE")
print("=" * 60)

print(
    final_dataset[
        [
            "material",
            "Cleaned_Formula",
            "Normalized_Formula"
        ]
    ].head(20)
)

print("\n")
print("=" * 60)
print("FINAL SHAPE")
print("=" * 60)

print(final_dataset.shape)

print("\nDataset Saved Successfully")

# ============================================================
# TEST CASES
# ============================================================

print("\n")
print("=" * 60)
print("TEST CASES")
print("=" * 60)

test_cases = [
    "Cu2ZnSnS4(CZTS)",
    "CZTSe",
    "CZTS",
    "Mg2Sn0.93Pb0.05Sb0.02 (engineering ZT)",
    "Bi2Te3 (p-type)",
    "ABC",
    "FeSe",
]

for t in test_cases:

    cleaned = clean_formula(t)

    print(f"\nINPUT:  {t}")
    print(f"CLEAN:  {cleaned}")

    if cleaned:
        print(
            f"VALID:  {strict_formula_check(cleaned)}"
        )
```
